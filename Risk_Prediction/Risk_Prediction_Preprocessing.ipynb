```python
# Identify duplicate rows
duplicates = df[df.duplicated()]
print("Duplicate Rows:\n", duplicates)

# Remove duplicate rows (keeping the first occurrence)
df = df.drop_duplicates()

# Check if duplicates are removed
print("Data after removing duplicates:\n", df)
```

```python
# Remove S.No. column from data
df = df.drop(['prognosis','Dosha','Risk for prognosis','Medicine', 'Unnamed: 78'], axis = 1)
```

```python
df.isnull().sum()
```

```python
df = df.copy()
df['Sweating'] = df['Sweating'].fillna(0).astype(int)
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
# Find unique values in the 'Name' column
unique_names = df['Risk'].unique()

print(unique_names)
```

```python
# Replace 'High Risk' with 'High' in the 'Risk' column
df['Risk'] = df['Risk'].replace('High Risk', 'High')

print(df)
```

```python
# After cleaning the dataset, save it to a new CSV file
df.to_csv('cleaned_dataset.csv', index=False)
```
